## Multi Linear Regression
Multi Linear Regression is an extension of the Linear Regression.

We had clear idea about how simple linear regression works. Now as we mean there will be more independent variables.
But adding more independent variables to a multiple regression procedure does not mean the regression will be "better" or offer better predictions; in fact it make thing worse. This is called overfitting.

The addition of more independent variables creates more relationships among them. So not only are the independent variables potentially related to the dependent variable, they are also potentially related to each other.When this happens, it is called multicollinearity.

The ideal condition is for all of the independent variables to be correlated with the dependent variable but NOT with each other.

Due to these reasons there is a fair amount of prep-work to BEFORE conducting multiple regression analysis if one is to do it properly.

## We will see the data and their preparation work.

```python
x1 = [89,66,78,111,44,77,80,66,100,76]
x2 = [4,1,3,6,1,3,3,2,5,3]
x3=[3.84,3.19,3.78,3.89,3.57,3.57,3.03,3.51,3.54,3.24]
y = [7,5.4,6.6,7.4,4.8,6.4,7,5.6,7.3,6.4]
```

Here x1 and x2 are the independent variables and y is dependent variable.

For Multi Linear Regression the formula would be 
$$y' = {b0}+{b1}x_1+{b2}x_2+ ... +{bp}x_p$$

As per the earlier discussion we need to see that there should be linear relationship between Independent and dependent 
variables to overcome overfitting.
In our example we can see that x1 and y, x2 and y have relationship where as x3 and y doesn't
There should be no linear realtionship between independent variables.
In our example we can see that x1 and x2 have relationship where as x2 and x3, x1 and x3 doesn't.
So bit a tussle which variables to use for the best model. Let us try all permutation and combination of variables and check which will be best.

```python
import pandas as pd
def linearModel(X,y):
    from sklearn.linear_model import LinearRegression  
    regressor = LinearRegression()  
    regressor.fit(X, y)
    y_pred = regressor.predict(X) 
    from sklearn import metrics  
    print('Mean Absolute Error:', metrics.mean_absolute_error(y, y_pred))  
    print('Mean Squared Error:', metrics.mean_squared_error(y, y_pred))  
    print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y, y_pred)))
    print('Co-efficients :',regressor.coef_)
    print('Intercept :',regressor.intercept_)
```

```python

df = pd.DataFrame({'x1':x1,'y':y})
X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values  
linearModel(X,y)
```

```python
df = pd.DataFrame({'x2':x2,'y':y})
X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values  
linearModel(X,y)
```

```python
df = pd.DataFrame({'x3':x3,'y':y})
X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values  
linearModel(X,y)
```

## You can see that I have calculated the Mean Absolute Error, Mean square error and Root Mean Squared Error and this was about the simple linear regression among x1,x2,x3 and y. Lesser the Mean square error better the model . In all these three model I could find that 1st Model was better as it had lesser error rate. We work on multiple variables now. 

```python
df = pd.DataFrame({'x1':x1,'x2':x2,'y':y})
X = df.iloc[:, :2].values  
y = df.iloc[:, 2].values  
linearModel(X,y)
```

```python
df = pd.DataFrame({'x2':x2,'x3':x3,'y':y})
X = df.iloc[:, :2].values  
y = df.iloc[:, 2].values  
linearModel(X,y)
```

```python
df = pd.DataFrame({'x1':x1,'x3':x3,'y':y})
X = df.iloc[:, :2].values  
y = df.iloc[:, 2].values  
linearModel(X,y)
```

```python
df = pd.DataFrame({'x1':x1,'x2':x2,'x3':x3,'y':y})
X = df.iloc[:, :3].values  
y = df.iloc[:, 3].values  
linearModel(X,y)
```

#### Now we have around 7 models here. The question arises which one to choose our best model. We see that first 2 linear regressions are good enough. But the third one is not as good as the first two. 

Now we see the fourth model which has 2 independent variables which are highly corelated. This model is providing better results. But which one is better performer in the independent variables has become the tragedy here. Therefore considering this model is bit tricky.

Fifth, sixth and seventh model has the negative coefficient which will not suit to our model.
